```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('housing.data.txt',header=None,delim_whitespace=True)
data.head()
```

```python
data.columns=['CRIM','ZN','INDUS','CCR','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
data.head()
```

```python
data.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
param = [{'n_neighbors':[i for i in range(1,11)],'weights':['uniform']},
        {'n_neighbors':[i for i in range(1,11)],'weights':['distance']}]
cv = GridSearchCV(estimator=KNeighborsRegressor(),param_grid=param,cv=KFold(n_splits=5))

scaler = StandardScaler(copy=True,with_mean=True,with_std=True).fit_transform(data)
X = scaler[:,:13]
Y = scaler[:,-1]

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)

cv.fit(X_train,Y_train)
```

```python
cv.best_estimator_
```

```python
knn = KNeighborsRegressor(n_neighbors=4,weights='distance')

knn.fit(X_train,Y_train)
```

```python
knn.score(X_test,Y_test)
```
